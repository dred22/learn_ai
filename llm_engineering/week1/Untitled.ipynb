```python
        options = Options()
        #if headless:
        options.add_argument("--headless")  # Запуск браузера без графического интерфейса
        options.add_argument("--no-sandbox")
        options.add_argument("--disable-dev-shm-usage")
        options.add_argument(f"user-agent={headers['User-Agent']}") # Передача User-Agent

        # Укажите путь к вашему chromedriver.exe, если он не в PATH
        # service = Service(executable_path='C:/path/to/your/chromedriver.exe')
        # driver = webdriver.Chrome(service=service, options=options)
        driver = webdriver.Chrome(options=options) # Если chromedriver в PATH
        try:
            driver.get("https://fr.allyz.com")

            # Добавьте явное ожидание для загрузки контента.
            # Это ОЧЕНЬ ВАЖНО для SPA.
            # Можно ждать появления определенного элемента,
            # или просто дать фиксированное время на загрузку.
            # В данном примере ждем, пока body не будет виден.
            # Вы можете настроить это под конкретный сайт.
            WebDriverWait(driver, 1).until(
                EC.presence_of_element_located((By.TAG_NAME, "body"))
            )
            # Иногда полезно добавить небольшую фиксированную задержку,
            # если контент загружается асинхронно после появления DOM
            time.sleep(1) # Например, подождать еще 1 секунду после загрузки DOM

            # Получаем весь HTML после рендеринга JavaScript
            rendered_html = driver.page_source
            soup = BeautifulSoup(rendered_html, 'html.parser')
            for irrelevant in soup.body(["script", "style", "img", "input"]):
                irrelevant.decompose()
            #self.text = soup.body.get_text(separator="\n", strip=True)
            print(soup.body.get_text(separator="\n", strip=True))
            #self._parse_soup(soup)

        except Exception as e:
            print(f"Error fetching {self.url} with Selenium: {e}")
            self.text = f"Error: Could not fetch page content with Selenium. {e}"
        finally:
            driver.quit() # Важно закрыть браузер после использования
```
